```python
import pandas as pd
import json

url = "https://data.cityofchicago.org/resource/qhb4-qx8k.csv"
data = pd.read_csv(url)

date_columns = ['start_date', 'end_date', 'modified_date']
data[date_columns] = data[date_columns].apply(pd.to_datetime)

def parse_location_json(location_str):
    if pd.isna(location_str):
        return None
    try:
        location_dict = json.loads(location_str.replace("'", '"'))
        return location_dict['coordinates']
    except (json.JSONDecodeError, KeyError):
        return None

data['location_coordinates'] = data['location'].apply(parse_location_json)
data.drop(columns=['location'], inplace=True)

unnecessary_columns = ['@computed_region_rpca_8um6', '@computed_region_vrxf_vc4k',
                      '@computed_region_6mkv_f3dw', '@computed_region_bdys_3d7i',
                      '@computed_region_43wa_7qmu', '@computed_region_awaf_s7ux']

for column in unnecessary_columns:
    try:
        data.drop(columns=[column], inplace=True)
    except KeyError:
        pass

data['notification_type'] = data['notification_type'].astype('category')
data['street_number'] = pd.to_numeric(data['street_number'], errors='coerce')

print(data.info())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://data.cityofchicago.org/resource/qhb4-qx8k.csv"
data = pd.read_csv(url)

data['start_date'] = pd.to_datetime(data['start_date'])

data['year'] = data['start_date'].dt.year

notifications_per_year_type = data.groupby(['year', 'notification_type']).size().unstack(fill_value=0)

notifications_per_year_type.plot(kind='line', marker='o', figsize=(12, 8))
plt.title('Asbestos Notifications Trends by Year and Type')
plt.xlabel('Year')
plt.ylabel('Number of Notifications')
plt.grid(True)
plt.legend(title='Notification Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://data.cityofchicago.org/resource/qhb4-qx8k.csv"
data = pd.read_csv(url)

location_counts = data.groupby('address').size().reset_index(name='notification_count')
location_counts = location_counts.sort_values(by='notification_count', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(location_counts['address'][:10], location_counts['notification_count'][:10])
plt.title('Top 10 Locations with the Most Notifications')
plt.xlabel('Location')
plt.ylabel('Notification Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://data.cityofchicago.org/resource/qhb4-qx8k.csv"
data = pd.read_csv(url)

location_counts = data.groupby('address').size().reset_index(name='notification_count')


top_10_locations = location_counts.nlargest(10, 'notification_count')['address']


filtered_data = data[data['address'].isin(top_10_locations)]

filtered_data['start_date'] = pd.to_datetime(filtered_data['start_date'])


notifications_by_year_location = filtered_data.groupby(['address', filtered_data['start_date'].dt.year]).size().unstack()

plt.figure(figsize=(12, 8))
notifications_by_year_location.plot(kind='bar', stacked=False)
plt.title('Notifications by Year for Top 10 Locations')
plt.xlabel('Location')
plt.ylabel('Notification Count')
plt.xticks(rotation=45)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import folium

url = "https://data.cityofchicago.org/resource/qhb4-qx8k.csv"
data = pd.read_csv(url)

data['start_date'] = pd.to_datetime(data['start_date'])

plt.figure(figsize=(10, 6))
plt.scatter(data['longitude'], data['latitude'], s=5, alpha=0.5)
plt.title('Spatial Distribution of Asbestos and Demolition Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

notifications_by_year = data['start_date'].dt.year.value_counts().sort_index()
plt.figure(figsize=(10, 6))
notifications_by_year.plot(kind='bar')
plt.title('Trends in Notifications Over Time')
plt.xlabel('Year')
plt.ylabel('Notification Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

map_chicago = folium.Map(location=[data['latitude'].mean(), data['longitude'].mean()], zoom_start=10)
heatmap_data = data[['latitude', 'longitude']].dropna().values.tolist()
map_chicago.save('notifications_heatmap.html')
```
